### **1. Imports**

```python
import joblib
from sklearn.pipeline import Pipeline
import pandas as pd
```

### **2. Load the Pipeline**

```python
loaded_model=joblib.load("../models/revenue_prediction_pipeline.pkl")
print("model loaded successfully")
```

### **3. Simulate New Business Data** 

```python
new_data = pd.DataFrame({
    "marketing_spend": [5000],
    "website_visits": [20000],
    "conversion_rate": [0.05],
    "num_customers": [1000],
    "avg_order_value": [50],
    "discount_rate": [0.10],
    "season": ["Summer"]
})

prediction = loaded_model.predict(new_data)

print("Predicted Revenue : ", prediction[0])
```
